```python
# Dijkstra's algorithm class
class Dijkstra :

    #intialization
    def __init__(self):
        self.graph = dict()
        self.costs = dict()
        self.parents = dict()
        self.processed = []
        self.start = 'start'

    # Casting
    def casting(self, casted_value):
        try :
            if '.' in casted_value :
                return float(casted_value)
            else :
                return int(casted_value)    
        except :       
            return casted_value.lower()

    # Insert edge
    def insert_edge(self, u, v, weight):
        u = self.casting(u)
        v = self.casting(v)
        weight = self.casting(weight)
        if not u in self.graph :
            self.graph[u] = {}
        self.graph[u][v] = weight

    # My graph
    def my_graph(self, sources, destinations, weights):
        number_of_nodes = len(sources)
        destinations = ['finish' if i == 'fin' else i for i in destinations]
        for i in range(number_of_nodes):
            self.insert_edge(sources[i], destinations[i], weights[i])               

    # Find lowest cost node
    def find_lowest_cost_node(self, costs):
        lowest_cost = float("inf")
        lowest_cost_node = None
        for node in costs:
            cost = costs[node]
            if cost < lowest_cost and node not in self.processed :
                lowest_cost = cost 
                lowest_cost_node = node
        return lowest_cost_node

    # Update tables
    def update_tables(self, sources, destinations, weights): 
        self.my_graph(sources, destinations, weights)
        start = self.start
        start = self.casting(start)
        if start in self.graph :
            self.costs[start] = 0
            self.parents[start] = None
            node = self.find_lowest_cost_node(self.costs)
            while node is not None :
                cost = self.costs[node]
                neighbors  = self.graph.get(node, {})
                for n in neighbors.keys():
                    new_cost = cost + neighbors[n]
                    if n not in self.costs or self.costs[n] > new_cost :
                        self.costs[n] = new_cost
                        self.parents[n] = node
                self.processed.append(node)
                node = self.find_lowest_cost_node(self.costs) 
            return self.costs, self.parents
        else :
            return None

    # Minimum cost
    def min_cost(self, costs_table):
        return costs_table['finish']   

    # Shortest path
    def shortest_path(self, parents_table):
        lst_path = []
        print('\nThe path is : ',end="")
        node = 'finish'
        while node in parents_table and parents_table[node] != 'start' :
            lst_path.append(node)
            node = parents_table[node]
        lst_path.reverse()
        lst_path.insert(0, node)
        node = parents_table[node]
        lst_path.insert(0, node)
        for i in range(len(lst_path)):
            if i == len(lst_path)-1 :
                print(f'{lst_path[i]}',end="")
            else :    
                print(f'{lst_path[i]} -> ',end="")         

    # Find shortest path
    def find_shortest_path(self, sources, destinations, weights):
        shortest_path = self.update_tables(sources, destinations, weights)
        print('\nCosts table -> ')
        costs_table = shortest_path[0]
        for k,v in costs_table.items():
            print(f'{k} : {v}')   
        parents_table = shortest_path[1]  
        print('\nParents table -> ')
        for k,v in parents_table.items():
            print(f'{k} : {v}')  
        print(f'\nThe shortest path cost is : {self.min_cost(costs_table)}')  
        self.shortest_path(parents_table)  
       
if __name__ == "__main__":
    number_of_edges = int(input('Enter number of edges : '))
    sources, destinations, weights = [],[],[]
    for i in range(number_of_edges):
        u, v, w = input(f'For edge {i+1}, enter source, destination and weight : ').split()
        sources.append(u)
        destinations.append(v)
        weights.append(w)   
    dijkstra = Dijkstra()
    dijkstra.find_shortest_path(sources, destinations, weights)
```
