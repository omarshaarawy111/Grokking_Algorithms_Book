```python
# Hash table class
class HashTable:

    # Initialization
    def __init__(self):
        self.dic = dict()
        
    # Casting
    def casting(self, casted_value):
        try:
            if '.' in casted_value:
                return float(casted_value)
            else:
                return int(casted_value)
        except:
            return casted_value

    # Hash table
    def insert_items(self, keys, values):
        if len(keys) == len(values):
            for i in range(len(keys)):
                key = keys[i]
                value = values[i]
                # if key in self.dic.keys():
                if self.dic.get(key):
                    self.dic[key] = value
                key = self.casting(key)
                value = self.casting(value)
                self.dic[key] = value
        else:
            return False

    # Display hash table
    def display_hash_table(self):
        if len(self.dic.keys()) == 0:
            print('Empty hash table.')
        else:
            self.dic = dict(sorted(self.dic.items()))
            print(f'Hash table : {self.dic}\n')
            for k, v in self.dic.items():
                print(f'Key {k}, Type {type(k)} and value {v}, Type {type(v)}.')

if __name__ == "__main__":
    keys = list(input('Enter keys : ').split())
    values = list(input('Enter Values : ').split())
    hash_table = HashTable()
    hash_table.insert_items(keys, values)
    hash_table.display_hash_table()
```

```python
# Cache class
class Cache:
    
    # Initialization
    def __init__(self):
        self.cache = dict()

    # Cache table
    def insert_items(self, urls, data):
        if len(urls) == len(data):
            for i in range(len(urls)):
                url = urls[i].lower()
                single_data = data[i].capitalize()
                # If url in self.cache.keys():
                if self.cache.get(url):
                    self.cache[url] = single_data
                self.cache[url] = single_data
        else:
            return False
        
    # Display cache table
    def display_cache_table(self):
        if len(self.cache.keys()) == 0:
            return 'Empty cache table.'
        else:
            self.cache = dict(sorted(self.cache.items()))
            return f'Cache table : {self.cache}\n'
     
    # Server 
    def get_data_from_server(self, url):
         return str(url).split('.com')[0]
    
    # page data
    def get_page(self, url):
            url = url.lower()
            if self.cache.get(url):
                 return self.cache[url]
            else:
                data = self.get_data_from_server(url)
                self.cache[url] = data
                return data

if __name__ == "__main__":
    urls = list(input('Enter keys : ').split())
    data = list(input('Enter Values : ').split())
    cache = Cache()
    cache.insert_items(urls, data)
    print(cache.display_cache_table())
    new_urls = list(input('Enter new requests').split())
    for i in new_urls:
        cache.get_page(i)
    print(cache.display_cache_table())
```

```python
import time

# Hash table with collision handling class
# Here we use chaining method
class HashTableChaining:

    # Initialization
    def __init__(self, size=100, max_load_factor=0.75):
        self.size = max(1, size)
        self.max_load_factor = max_load_factor
        self.occupied_slots = 0
        self.hash_table = [[] for i in range(self.size)]

    # Casting
    def casting(self, casted_value):
        try:
            if isinstance(casted_value, str) and '.' in casted_value:
                return float(casted_value)
            else:
                return int(casted_value)
        except:
            return casted_value

    # Hash function
    def hash_function(self, key):
        # This is the python built in function of hash
        # It is the best as it is good uniformly distributed keys evenly
        # Here you return the hash value
        return hash(key) % self.size

    # Load Factor function
    def load_factor(self):
        return self.occupied_slots / self.size

    # Resize function
    def _resize(self):
        old_table = self.hash_table
        self.size *= 2
        # We create new table and insert elements from old one
        self.hash_table = [[] for i in range(self.size)]
        # Make it zero
        self.occupied_slots = 0
        for chain in old_table:
            for key, value in chain:
                self.insert_single(key, value)

    # Insert single function
    def insert_single(self, key, value):
        # For inserting new or old
        index = self.hash_function(key)
        # Here you loop for every index which is a list of [key, value] pairs
        for pair in self.hash_table[index]:
            if pair[0] == key:
                # Update existing key
                pair[1] = value
                return

        # If key does not exist, append new pair
        self.hash_table[index].append([key, value])
        self.occupied_slots += 1

    # Insert (collision handling)
    def insert(self, keys, values):
        if len(keys) == len(values):
            for i in range(len(keys)):
                key = keys[i]
                value = values[i]
                key = self.casting(key)
                value = self.casting(value)

                # Threshold of load factor
                if (self.occupied_slots + 1) / self.size > self.max_load_factor:
                    # Resize first
                    self._resize()

                self.insert_single(key, value)
        else:
            return False

    # Search
    def search(self, key):
        key = self.casting(key)
        index = self.hash_function(key)
        for pair in self.hash_table[index]:
            if pair[0] == key:
                value = pair[1]
                print(f'The value of the key is {value}.')
                return value
        print('Key not found.')
        return None

    # Delete
    def delete(self, key):
        key = self.casting(key)
        index = self.hash_function(key)
        for i, pair in enumerate(self.hash_table[index]):
            if pair[0] == key:
                del self.hash_table[index][i]
                # Decrease occupied slots by one
                self.occupied_slots -= 1
                print(f'Key {key} successfully deleted.')
                return
        print('Key not found.')

    # Display table
    def display_table(self):
        print('Hash table : ')
        print(f'Current size : {self.size}')
        print(f'Number of elements : {self.occupied_slots}')
        print(f'Load factor : {self.load_factor():.3f}')
        for i, pair in enumerate(self.hash_table):
            print(f'slot {i} : {pair}')

if __name__ == "__main__":
    size = int(input('Enter the size of the hash table : '))
    print(f'Size : {size}')
    keys = list(input('Enter keys : ').split())
    values = list(input('Enter values : ').split())
    # Keep load factor as standard
    hash_table = HashTableChaining(size)
    hash_table.insert(keys, values)
    hash_table.display_table()
    time.sleep(2)
    searched_key = input('Enter the key you are looking for : ')
    hash_table.search(searched_key)
    deleted_key = input('Enter the key you want to delete : ')
    hash_table.delete(deleted_key)
    hash_table.search(deleted_key)
    hash_table.display_table()
```
